```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = [ "DenseNet201" , "EfficientNetB7" , "MobileNetV2"]
Traning_loss =        [11.5,3.6,6.8]
Traning_accuracy  =   [97.6,98.9,98.5]
Valadation_loss=      [9.4,7.1,7.6]
Valadation_accuracy = [96.6,98.5,97.6]


x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars


fig, ax = plt.subplots()
rects1 = ax.bar(x +0.0, Traning_loss, width, label='Traning_loss(%)', color = "r")
rects2 = ax.bar(x + 0.25, Traning_accuracy, width, label='Traning_accuracy(%)',color = "b")
rects3 = ax.bar(x + 0.50, Valadation_loss, width, label='Validation_loss(%)', color = "orange")
rects4 = ax.bar(x +  0.75, Valadation_accuracy, width, label='Validation_accuracy(%)', color = "g")
rects5 = ax.bar(x +0.0, Traning_loss, width,  color = "r")
rects6 = ax.bar(x + 0.25, Traning_accuracy, width ,color = "b")
rects7 = ax.bar(x + 0.50, Valadation_loss, width, color = "orange")
rects8 = ax.bar(x +  0.75, Valadation_accuracy, width, color = "g")
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Percentage %')
ax.set_title('Comparison of DenseNet201 , EfficientNetB7 and MobileNetV2')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)
autolabel(rects5)
autolabel(rects6)
autolabel(rects7)
autolabel(rects8)

fig.tight_layout()

plt.show()
```
